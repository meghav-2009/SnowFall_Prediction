```python
import requests  ## for getting data from a server GET
import re   ## for regular expressions
import pandas as pd    ## for dataframes and related
from pandas import DataFrame
```

```python
endPoint = "http://api.weatherapi.com/v1/forecast.json?key=3d596aaf43d0484b90204048233101&q=80309&days=14"
# data related to syracuse city
```

```python
response=requests.get(endPoint)
print(response) 
```

```python
jsontxt = response.json()
jsontxt
```

```python
filename="weather_df.csv"
MyFILE=open(filename,"w")
WriteThis="Time,Temp_c,Wind_mph,Wind_degree,Wind_dir,Is_day,Condition,Precip_mm,Visib_miles,Humidity,Cloud,Feels_like_c,Dew_point_c,UV,Total_snow_cm,Chance_of_rain,Chance_of_snow\n"
MyFILE.write(WriteThis)
MyFILE.close()
```

```python
MyFILE=open(filename, "a")

for each in jsontxt['forecast']['forecastday']:
    
    Total_snow_cm = each['day']['totalsnow_cm']
    
    for i in range(0,24):
        
        Time = each['hour'][i]['time'].split()[1]
        Temp_c = each['hour'][i]['temp_c']
        Wind_mph = each['hour'][i]['wind_mph']
        Wind_degree = each['hour'][i]['wind_degree']
        Wind_dir = each['hour'][i]['wind_dir']
        Is_day = each['hour'][i]['is_day']
        Condition = each['hour'][i]['condition']['text']
        Precip_mm = each['hour'][i]['precip_mm']
        Visib_miles = each['hour'][i]['vis_miles']
        Humidity = each['hour'][i]['humidity']
        Cloud = each['hour'][i]['cloud']
        Feels_like_c = each['hour'][i]['feelslike_c']
        Dew_point_c = each['hour'][i]['dewpoint_c']
        UV = each['hour'][i]['temp_c']
        Chance_of_rain = each['hour'][i]['chance_of_rain']
        Chance_of_snow = each['hour'][i]['chance_of_snow']
        
        WriteThis = str(Time)+","+str(Temp_c)+","+ str(Wind_mph) + "," + str(Wind_degree) + ","+str(Wind_dir)+","+str(Is_day)+","+ str(Condition) + "," + str(Precip_mm) + ","+str(Total_snow_cm)+","+str(Visib_miles)+","+ str(Humidity) + "," + str(Cloud) + "," +str(Feels_like_c)+","+str(Dew_point_c)+","+str(UV)+","+ str(Chance_of_rain) +","+ str(Chance_of_snow) + "\n"
        
        MyFILE.write(WriteThis)
    
MyFILE.close()
```
